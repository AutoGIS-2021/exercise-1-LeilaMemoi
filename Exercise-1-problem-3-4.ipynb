````python
# REPLACE THE ERROR BELOW WITH YOUR OWN CODE
list1=[]
list2=[]
def create_od_lines(list1,list2):
    for start,end in zip(list1,list2):
        lines=LineString(start,end)
    return(lines)
    
    
````

````python
def calculate_total_distance(lines):
    tot_length=0
    for line in lines:
        distance=lines.length
        tot_length+=distance
        return tot_length
````
